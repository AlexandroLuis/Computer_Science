# Testando a base de textos

```python
dataset = pd.read_csv('Corona_NLP_train.csv',encoding='iso8859')
dataset
```

```python
dataset = dataset.dropna().sample(20000)
dataset.reset_index(inplace=True,drop=True)
dataset
```

# BERTopic

```python
topic_model = BERTopic(embedding_model="distiluse-base-multilingual-cased-v1", calculate_probabilities=True, verbose=True)
topics, probs = topic_model.fit_transform(dataset.OriginalTweet.to_list())
```

```python
topic_model.get_topic_info().head(50)
```

```python
topic_model.visualize_barchart(topics=[0,1,2,3,4,5])
```
